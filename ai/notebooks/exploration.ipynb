```python
# Bloc 1 : Importer les bibliothèques nécessaires
# Ce code charge les outils dont nous avons besoin

import pandas as pd          # Pour manipuler les données (tableaux)
import numpy as np           # Pour les calculs mathématiques
import matplotlib.pyplot as plt  # Pour créer des graphiques
import seaborn as sns        # Pour de beaux graphiques
from datetime import datetime    # Pour gérer les dates

# Configuration pour de beaux graphiques
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("✅ Bibliothèques chargées avec succès !")
print("=" * 60)
```

```python
# Bloc 2 : Charger les 4 datasets CSV
# Pandas va lire les fichiers CSV et les transformer en tableaux

print("📁 Chargement des datasets...")
print()

# Dataset 1 : Génération Plant 1
df_gen1 = pd.read_csv('../datasets/Plant_1_Generation_Data.csv')
print(f"✅ Plant 1 Generation : {df_gen1.shape[0]} lignes, {df_gen1.shape[1]} colonnes")

# Dataset 2 : Génération Plant 2
df_gen2 = pd.read_csv('../datasets/Plant_2_Generation_Data.csv')
print(f"✅ Plant 2 Generation : {df_gen2.shape[0]} lignes, {df_gen2.shape[1]} colonnes")

# Dataset 3 : Météo Plant 1
df_weather1 = pd.read_csv('../datasets/Plant_1_Weather_Sensor_Data.csv')
print(f"✅ Plant 1 Weather : {df_weather1.shape[0]} lignes, {df_weather1.shape[1]} colonnes")

# Dataset 4 : Météo Plant 2
df_weather2 = pd.read_csv('../datasets/Plant_2_Weather_Sensor_Data.csv')
print(f"✅ Plant 2 Weather : {df_weather2.shape[0]} lignes, {df_weather2.shape[1]} colonnes")

print()
print("=" * 60)
print("🎉 TOUS LES DATASETS SONT CHARGÉS !")
print("=" * 60)
```

```python
# Bloc 3 : Regarder les premières lignes des données
# Cela nous permet de voir à quoi ressemblent nos données

print("👀 APERÇU DES DONNÉES")
print("=" * 60)
print()

print("📊 Plant 1 - Generation Data (5 premières lignes) :")
print(df_gen1.head())
print()

print("Colonnes disponibles :", list(df_gen1.columns))
print()

print("-" * 60)
print()

print("🌤️ Plant 1 - Weather Data (5 premières lignes) :")
print(df_weather1.head())
print()

print("Colonnes disponibles :", list(df_weather1.columns))
print()

print("=" * 60)
```

```python
# Bloc 4 : Statistiques descriptives
# Comprendre les valeurs : minimum, maximum, moyenne, etc.

print("📈 STATISTIQUES DESCRIPTIVES")
print("=" * 60)
print()

print("⚡ AC_POWER (Puissance générée) - Plant 1 :")
print(df_gen1['AC_POWER'].describe())
print()
print(f"   → Puissance minimale : {df_gen1['AC_POWER'].min():.2f} W")
print(f"   → Puissance maximale : {df_gen1['AC_POWER'].max():.2f} W")
print(f"   → Puissance moyenne : {df_gen1['AC_POWER'].mean():.2f} W")
print()

print("-" * 60)
print()

print("☀️ IRRADIATION (Rayonnement solaire) - Plant 1 :")
print(df_weather1['IRRADIATION'].describe())
print()
print(f"   → Irradiation minimale : {df_weather1['IRRADIATION'].min():.2f} W/m²")
print(f"   → Irradiation maximale : {df_weather1['IRRADIATION'].max():.2f} W/m²")
print(f"   → Irradiation moyenne : {df_weather1['IRRADIATION'].mean():.2f} W/m²")
print()

print("-" * 60)
print()

print("🌡️ TEMPÉRATURE AMBIANTE - Plant 1 :")
print(df_weather1['AMBIENT_TEMPERATURE'].describe())
print()
print(f"   → Température minimale : {df_weather1['AMBIENT_TEMPERATURE'].min():.2f} °C")
print(f"   → Température maximale : {df_weather1['AMBIENT_TEMPERATURE'].max():.2f} °C")
print(f"   → Température moyenne : {df_weather1['AMBIENT_TEMPERATURE'].mean():.2f} °C")
print()

print("=" * 60)
```

```python
# Bloc 5 : Vérifier s'il y a des valeurs manquantes
# Les valeurs manquantes sont des "trous" dans les données

print("🔍 DÉTECTION DES VALEURS MANQUANTES")
print("=" * 60)
print()

def verifier_manquantes(df, nom):
    """Fonction pour vérifier les valeurs manquantes dans un dataset"""
    total = df.isnull().sum().sum()  # Nombre total de valeurs manquantes
    
    print(f"📊 {nom}")
    if total > 0:
        print(f"   ⚠️  {total} valeurs manquantes détectées :")
        manquantes = df.isnull().sum()
        for colonne, nb in manquantes.items():
            if nb > 0:
                pourcentage = (nb / len(df)) * 100
                print(f"      - {colonne} : {nb} ({pourcentage:.2f}%)")
    else:
        print(f"   ✅ Aucune valeur manquante !")
    print()

# Vérifier chaque dataset
verifier_manquantes(df_gen1, "Plant 1 - Generation")
verifier_manquantes(df_gen2, "Plant 2 - Generation")
verifier_manquantes(df_weather1, "Plant 1 - Weather")
verifier_manquantes(df_weather2, "Plant 2 - Weather")

print("=" * 60)
```

```python
# CORRECTION : Créer les dossiers nécessaires
# Ce code crée automatiquement les dossiers s'ils n'existent pas

import os

print("🔧 CRÉATION DES DOSSIERS NÉCESSAIRES...")
print()

# Liste des dossiers à créer
dossiers = [
    '../schemas',
    '../preprocessing',
    '../models',
    '../training',
    '../inference'
]

# Créer chaque dossier
for dossier in dossiers:
    if not os.path.exists(dossier):
        os.makedirs(dossier)
        print(f"✅ Dossier créé : {dossier}")
    else:
        print(f"ℹ️  Dossier existe déjà : {dossier}")

print()
print("=" * 60)
print("🎉 TOUS LES DOSSIERS SONT PRÊTS !")
print("=" * 60)
```

```python
# Bloc 6 CORRIGÉ : Créer des graphiques pour visualiser les données
# Les graphiques nous aident à VOIR les patterns dans les données

print("📊 CRÉATION DES GRAPHIQUES...")
print()

# D'abord, convertir les dates en format datetime
# dayfirst=True car les dates sont au format jour-mois-année
df_gen1['DATE_TIME'] = pd.to_datetime(df_gen1['DATE_TIME'], dayfirst=True)
df_weather1['DATE_TIME'] = pd.to_datetime(df_weather1['DATE_TIME'], dayfirst=True)

print("✅ Dates converties au format datetime")
print(f"   Période des données : {df_gen1['DATE_TIME'].min()} à {df_gen1['DATE_TIME'].max()}")
print()

# Créer une grande figure avec 4 graphiques
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Haské Énergie - Analyse des Données Solaires (Plant 1)', 
             fontsize=18, fontweight='bold', y=0.995)

# ---- Graphique 1 : AC_POWER dans le temps ----
# On prend les 2000 premières valeurs pour que ce soit lisible
axes[0, 0].plot(df_gen1['DATE_TIME'].head(2000), 
                df_gen1['AC_POWER'].head(2000), 
                color='darkorange', linewidth=1, alpha=0.8)
axes[0, 0].set_title('Production d\'Énergie (AC_POWER) au fil du temps', 
                     fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Date et Heure', fontsize=10)
axes[0, 0].set_ylabel('Puissance (W)', fontsize=10)
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].tick_params(axis='x', rotation=45)

# ---- Graphique 2 : Distribution de AC_POWER ----
axes[0, 1].hist(df_gen1['AC_POWER'], bins=60, color='green', 
                alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Distribution de la Production (AC_POWER)', 
                     fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Puissance (W)', fontsize=10)
axes[0, 1].set_ylabel('Fréquence', fontsize=10)
axes[0, 1].grid(True, alpha=0.3, axis='y')

# ---- Graphique 3 : IRRADIATION dans le temps ----
axes[1, 0].plot(df_weather1['DATE_TIME'].head(2000), 
                df_weather1['IRRADIATION'].head(2000), 
                color='red', linewidth=1, alpha=0.8)
axes[1, 0].set_title('Irradiation Solaire au fil du temps', 
                     fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Date et Heure', fontsize=10)
axes[1, 0].set_ylabel('Irradiation (W/m²)', fontsize=10)
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].tick_params(axis='x', rotation=45)

# ---- Graphique 4 : TEMPÉRATURE dans le temps ----
axes[1, 1].plot(df_weather1['DATE_TIME'].head(2000), 
                df_weather1['AMBIENT_TEMPERATURE'].head(2000), 
                color='blue', linewidth=1, alpha=0.8)
axes[1, 1].set_title('Température Ambiante au fil du temps', 
                     fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Date et Heure', fontsize=10)
axes[1, 1].set_ylabel('Température (°C)', fontsize=10)
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].tick_params(axis='x', rotation=45)

# Ajuster l'espacement entre les graphiques
plt.tight_layout()

# Sauvegarder le graphique (maintenant le dossier existe !)
plt.savefig('../schemas/exploration_graphs.png', dpi=300, bbox_inches='tight')

print("✅ Graphiques créés avec succès !")
print("✅ Sauvegardés dans : ai/schemas/exploration_graphs.png")
print()

# Afficher les graphiques
plt.show()

print("=" * 60)
print("🎉 VISUALISATION TERMINÉE !")
print("=" * 60)
```

```python
# Bloc 7 : Analyse des corrélations
# Comprendre comment AC_POWER est lié à l'irradiation et la température

print("🔗 ANALYSE DES CORRÉLATIONS")
print("=" * 60)
print()

# Fusionner les datasets Generation et Weather pour Plant 1
# On fusionne sur la colonne DATE_TIME
df_merged = pd.merge(df_gen1, df_weather1, on='DATE_TIME', how='inner')

print(f"✅ Données fusionnées : {df_merged.shape[0]} lignes")
print(f"   Colonnes disponibles : {list(df_merged.columns)}")
print()

# Afficher un aperçu des données fusionnées
print("👀 Aperçu des données fusionnées (5 premières lignes) :")
print(df_merged[['DATE_TIME', 'AC_POWER', 'IRRADIATION', 'AMBIENT_TEMPERATURE']].head())
print()

print("-" * 60)
print()

# Calculer la matrice de corrélation
correlation = df_merged[['AC_POWER', 'IRRADIATION', 'AMBIENT_TEMPERATURE']].corr()

print("📊 MATRICE DE CORRÉLATION :")
print(correlation)
print()

# Interpréter les corrélations
print("💡 INTERPRÉTATION :")
corr_irrad = correlation.loc['AC_POWER', 'IRRADIATION']
corr_temp = correlation.loc['AC_POWER', 'AMBIENT_TEMPERATURE']

print(f"   🌞 AC_POWER ↔ IRRADIATION : {corr_irrad:.3f}")
if corr_irrad > 0.7:
    print(f"      → Corrélation FORTE et POSITIVE")
    print(f"      → Plus il y a de soleil, plus on produit d'énergie !")
elif corr_irrad > 0.4:
    print(f"      → Corrélation MODÉRÉE et positive")
else:
    print(f"      → Corrélation FAIBLE")

print()
print(f"   🌡️  AC_POWER ↔ TEMPÉRATURE : {corr_temp:.3f}")
if abs(corr_temp) > 0.7:
    print(f"      → Corrélation FORTE")
elif abs(corr_temp) > 0.4:
    print(f"      → Corrélation MODÉRÉE")
else:
    print(f"      → Corrélation FAIBLE")

print()
print("-" * 60)
print()

# Visualiser la matrice de corrélation
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(correlation, annot=True, fmt='.3f', cmap='coolwarm', 
            center=0, square=True, linewidths=2, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1, annot_kws={'size': 14, 'weight': 'bold'})

ax.set_title('Matrice de Corrélation - Haské Énergie (Plant 1)', 
             fontsize=16, fontweight='bold', pad=20)

plt.tight_layout()
plt.savefig('../schemas/correlation_matrix.png', dpi=300, bbox_inches='tight')

print("✅ Matrice de corrélation sauvegardée dans : ai/schemas/correlation_matrix.png")

plt.show()

print()
print("=" * 60)
print("✅ ANALYSE DE CORRÉLATION TERMINÉE !")
print("=" * 60)
print()

# Sauvegarder le dataset fusionné pour plus tard
df_merged.to_csv('../preprocessing/merged_plant1_data.csv', index=False)
print("💾 Dataset fusionné sauvegardé dans : ai/preprocessing/merged_plant1_data.csv")
print()

print("=" * 60)
print("🎉 EXPLORATION COMPLÈTE TERMINÉE !")
print("=" * 60)
```
